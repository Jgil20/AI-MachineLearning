# Random Forest Regression

## Importing the libraries

```
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv("Position_Salaries.csv")
Level = dataset.iloc[:,1:-1].values
Salary = dataset.iloc[:,-1].values
```

## Training the Random Forest Regression model on the whole dataset

```
from sklearn.ensemble import RandomForestClassifier
regressor = RandomForestClassifier(n_estimators = 10,random_state=0)
regressor.fit(Level,Salary)
```

## Predicting a new result

```
regressor.predict([[6.5]])
```

## Visualising the Random Forest Regression results (higher resolution)

```
Level_grid = np.arange(min(Level),max(Level),0.1)
Level_grid = Level_grid.reshape(len(Level_grid),1)
plt.scatter(Level,Salary,color="red")
plt.plot(Level_grid,regressor.predict(Level_grid),color="green")
plt.title("Decision Tree Regression")
plt.xlabel("Years in Position")
plt.ylabel("Salary")
plt.show()
```
