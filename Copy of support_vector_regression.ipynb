# Support Vector Regression (SVR)

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv("Position_Salaries.csv")
Level = dataset.iloc[:, 1:-1].values
Salary = dataset.iloc[:, -1].values
```

```
print(Level)
```

```
print(Salary)
```

```
Salary = Salary.reshape(len(Salary),1)
```

```
print(Salary)
```

## Feature Scaling

```
from sklearn.preprocessing import StandardScaler
standarScalerLevel = StandardScaler() 
standardScalerSalary = StandardScaler()
Level = standarScalerLevel.fit_transform(Level)
Salary = standardScalerSalary.fit_transform(Salary)
```

```
print(Level)
```

```
print(Salary)
```

## Training the SVR model on the whole dataset

```
from sklearn.svm import SVR
regressor = SVR(kernel = 'rbf')
regressor.fit(Level,np.ravel(Salary))
```

## Predicting a new result

```
standardScalerSalary.inverse_transform([regressor.predict(standarScalerLevel.transform([[6.5]]))])
```

## Visualising the SVR results

```
plt.scatter(standarScalerLevel.inverse_transform(Level), standardScalerSalary.inverse_transform(Salary), color = 'red')
plt.plot(standarScalerLevel.inverse_transform(Level), standardScalerSalary.inverse_transform([regressor.predict(Level)]).reshape(10,1), color = 'blue')
plt.title('Truth or Bluff (SVR)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```
from numpy.core.fromnumeric import reshape
plt.scatter(standarScalerLevel.inverse_transform(Level), standardScalerSalary.inverse_transform(Salary),color="green")
plt.plot(standarScalerLevel.inverse_transform(Level), standardScalerSalary.inverse_transform([regressor.predict(Level)]).reshape(10,1),color="blue")

plt.title("Truth or Bluff (SVR")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

## Visualising the SVR results (for higher resolution and smoother curve)

```
X_grid = np.arange(min(standarScalerLevel.inverse_transform(Level)), max(standarScalerLevel.inverse_transform(Level)), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(standarScalerLevel.inverse_transform(Level), standardScalerSalary.inverse_transform(Salary), color = 'red')
plt.plot(X_grid, standardScalerSalary.inverse_transform([regressor.predict(standarScalerLevel.transform(X_grid))]).reshape(90,1), color = 'blue')
plt.title('Truth or Bluff (SVR)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```
X_grid = np.arange(min(standarScalerLevel.inverse_transform(Level)), max(standarScalerLevel.inverse_transform(Level)),0.1)
X_grid = X_grid.reshape(len(X_grid),1)
plt.scatter(standarScalerLevel.inverse_transform(Level),standardScalerSalary.inverse_transform(Salary), color = "Red")
plt.plot(X_grid, standardScalerSalary.inverse_transform([regressor.predict(standarScalerLevel.transform(X_grid))]).reshape(90,1), color = "blue")
plt.title("Truth or Bluff (Linear Regression)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show();
```
